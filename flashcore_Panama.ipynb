```python
from IPython.display import clear_output
import time
from datetime import datetime, timedelta

def clean_screen():
    for i in range(5):
        clear_output(wait=True)  # Clears the previous output
#         print(f"Step {i + 1}")
#         time.sleep(1)
        
# live update 
# save_check_point('check_points/live.json', {'FOOTBALL': 'ARSENAL'})

def update_dict(dict_pending, time_update, delay, aux_list):
    current_time = datetime.now()
    if current_time > time_update:
        print("UPDATING DICT")
        for key_delete in aux_list:
            dict_pending.pop(key_delete)
        new_dict = load_check_point('check_points/live.json')
        
        if len(new_dict) != 0:
            for key, value in new_dict.items():                
                dict_pending[key] = value

        # Se calcula el siguiente tiempo de actualizacin
        return current_time + timedelta(seconds=delay)
    
    return time_update  # Si no hay actualizacin, se conserva el anterior

# Inicializar valores
dict_pending = load_check_point('check_points/live.json')
delay = 10  # segundos de espera entre actualizaciones
time_update = datetime.now() + timedelta(seconds=delay)
aux_list = []
# Bucle principal
while True:
    print(f"current time: {datetime.now()}")
    print(f" time_update: {time_update}")
    time_update = update_dict(dict_pending, time_update, delay, aux_list)    
    for sport_name, dict_match in dict_pending.items():
        if sport_name not in aux_list:
            print_section(f"{sport_name} {dict_match}", space_=10)
            aux_list.append(sport_name)
            print("aux_list: ", aux_list)
            time.sleep(1)
    time.sleep(1)

    stop_validate()
    clean_screen()
```

```python
SELECT 
    l.league_name AS LEAGUE_NAME,
    ct.country_name,
    COUNT(m.match_id) AS total_matches
FROM match m
JOIN season s ON m.season_id = s.season_id
JOIN league l ON s.league_id = l.league_id
JOIN sport sp ON l.sport_id = sp.sport_id
JOIN country ct ON l.country_id = ct.country_id
WHERE sp.name = 'FOOTBALL'
GROUP BY l.league_name,
ORDER BY total_matches DESC;


# SELECT LEAGUE_NAME USING TEAM_ID
# SELECCION DE PARTIDOS AGRUPADOS POR LIGA
SELECT m.match_id , md.match_detail_id    
FROM match_detail md
JOIN match m ON m.match_id = md.match_id
JOIN score_entity se ON md.match_detail_id = se.match_detail_id
WHERE md.team_id = '171fe704-08d9-4348-b02b-b1f227a2760e';



# SELECCION DE PARTIDOS AGRUPADOS POR LIGA
SELECT l.league_name AS LEAGUE_NAME,
    ct.country_name,
    COUNT(m.match_id) AS total_matches
FROM match m
JOIN season s ON m.season_id = s.season_id
JOIN league l ON s.league_id = l.league_id
JOIN sport sp ON l.sport_id = sp.sport_id
JOIN country ct ON l.country_id = ct.country_id
WHERE sp.name = 'FOOTBALL'
GROUP BY LEAGUE_NAME
ORDER BY total_matches DESC;

# SELECCION DE PARTIDOS AGRUPADOS POR SPORT NAME
SELECT sp.name AS sport_name, COUNT(m.match_id) AS total_matches
FROM match m
JOIN season s ON m.season_id = s.season_id
JOIN league l ON s.league_id = l.league_id
JOIN sport sp ON l.sport_id = sp.sport_id
GROUP BY sp.name
ORDER BY total_matches DESC;

# CONTENO DE LIGAS AGRUPADAS POR DEPORTE
SELECT sp.name AS sport_name, COUNT(l.league_id) AS total_leagues
FROM league l
JOIN sport sp ON l.sport_id = sp.sport_id
GROUP BY sp.name
ORDER BY total_leagues DESC;


# VERIFICACIOND EL NUMERO DE PARTIDOS CREADOS AGRUPADOS POR LIGAS
SELECT l.league_name, COUNT(m.match_id) AS total_matchs
FROM match m
JOIN season s ON m.season_id = s.season_id
JOIN league l ON s.league_id = l.league_id
JOIN sport sp ON l.sport_id = sp.sport_id
where sp.name = 'TENNIS'
GROUP BY l.league_name;

# VERIFICACIOND EL NUMERO DE PARTIDOS CREADOS AGRUPADOS POR LIGAS (incluyendo las de match cero)
SELECT 
    l.league_name,
    COUNT(m.match_id) AS total_matchs
FROM league l
JOIN sport sp ON l.sport_id = sp.sport_id AND sp.name = 'FOOTBALL'
LEFT JOIN season s ON l.league_id = s.league_id
LEFT JOIN match m ON s.season_id = m.season_id
GROUP BY l.league_name;

# MOSTRAR LAS LIGAS SEGUN EL DEPORTE
SELECT l.league_name
FROM league l
JOIN sport sp ON l.sport_id = sp.sport_id
where sp.name = 'FOOTBALL';

# QUERE PARA OBTENER INFORMACION RELACIONADA A MATCHS, SPORT, LEAGUES, COUNTRY


# GET REQUIRED INFORMATION ABOUT MATCH PENDING TO UPDATE
SELECT sp.name, c.country_name, l.league_name, m.match_date, m.match_id, m.status, m.name
FROM match m
JOIN league l ON m.league_id = l.league_id
JOIN sport sp ON l.sport_id = sp.sport_id
JOIN country c ON l.country_id = c.country_id
WHERE m.status = 'schedule' and m.match_date <=current_date;
```
